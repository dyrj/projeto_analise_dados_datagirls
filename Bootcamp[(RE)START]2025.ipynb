## Introdução ao Pandas

**O que é um DataFrame?**

Um DataFrame é a principal estrutura de dados da biblioteca Pandas. Ele funciona como uma **tabela de dados**, semelhante a uma planilha do Excel, com **linhas e colunas rotuladas**, onde cada coluna pode conter um tipo diferente de dado (números, textos, datas, etc.). É com o DataFrame que fazemos a maior parte das análises de dados em Python.

![](https://pandas.pydata.org/docs/_images/01_table_dataframe.svg)
<p><small>Aprofunde-se no assunto nesse <a href="https://pandas.pydata.org/docs/getting_started/intro_tutorials/01_table_oriented.html#min-tut-01-tableoriented">tutorial do Pandas</a>.</small></p>

**O que são Series no Pandas?**

Uma Series é uma estrutura de dados do Pandas que representa uma **coluna única de um DataFrame**. Ela é unidimensional e possui um índice, o que a torna parecida com uma lista ou vetor, mas com funcionalidades avançadas. A principal diferença é que enquanto o **DataFrame armazena várias colunas**, a Series representa apenas **uma coluna isolada**.

### Importar arquivos com Pandas

**Que tipos de arquivos podem se transformar em DataFrames no Pandas?**

O Pandas permite importar diversos tipos de arquivos e transformá-los em DataFrames, como **CSV, Excel, JSON, HTML, SQL e até arquivos Parquet**. Para isso, ele oferece várias funções específicas, como `read_csv()`, `read_excel()`, `read_json()`, entre outras. A função mais comum e utilizada em análises iniciais é a [`read_csv()`](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html), ideal para ler arquivos CSV.

![](https://pandas.pydata.org/docs/_images/02_io_readwrite.svg)
<p><small>Aprofunde-se no assunto nesse <a href="https://pandas.pydata.org/docs/getting_started/intro_tutorials/02_read_write.html#min-tut-02-read-write">tutorial do Pandas</a>.</small></p>

Além de importar arquivos, o Pandas também permite **exportar** DataFrames para outros formatos, como CSV (`to_csv()`), JSON (`to_json()`) e Excel (`to_excel()`), facilitando o compartilhamento ou uso dos dados em outras ferramentas.

## Manipualação de dados

### Filtragem e agrupamento de informações

Filtrando imóveis com preço abaixo de 250 mil reais e area menor que 25000

````python
df_filtrado = df[df['preco'] < 250000]
df_filtrado = df_filtrado[df_filtrado['area'] < 25000]
````

Observando a distribuição

````python
df_filtrado.plot.scatter(x='area', y='preco')
````

Agrupando e obtendo o preço médio por tipo de imóvel ([pandas.DataFrame.groupby](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html#pandas.DataFrame.groupby))

````python
df_filtrado.groupby('tipo')['preco'].mean()
````

### Aplicação de consultas simples

Consultando imóveis específicos ([pandas.DataFrame.query](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.query.html#pandas.DataFrame.query))

- Apenas Apartamentos que tem 2 quartos

````python
df_filtrado.query('tipo == "Apartamento" and quartos >= 2')
````

### Criação de novas colunas

Criando uma Series que calcula o custo por metro quadrado

````python
preco_area = df_filtrado['preco'] / df_filtrado['area']
preco_area
````

Criando coluna 'preco_por_area' que mostra no DataFrame o custo por metro quadrado

````python
df_filtrado['preco_por_area'] = preco_area
df_filtrado
````
